# Data Visualization

```python
plt.figure(figsize=(6,4))
sns.countplot(x = data['Vehicle_driver_relation'],color='red')
plt.title('Distribution of vehicle driver relation')
```

```python
weather_counts=data['Weather_conditions'].value_counts().reset_index()
weather_counts.columns=['Weather_conditions','Accidents']
sns.scatterplot(x='Weather_conditions',y='Accidents',data=weather_counts)
plt.title('Accidents by weather conditions')
plt.xticks(rotation=45)
plt.show()
```

```python
vehicle_accidents=data['Type_of_vehicle'].value_counts()
plt.figure(figsize=(12,8))
vehicle_accidents.plot(kind='barh', color='lightgreen')
plt.title('Accidents by vehicle')
plt.show()
```

```python
Road_surface_counts=data['Sex_of_driver'].value_counts()
plt.figure(figsize=(6,4))
plt.pie(Road_surface_counts,labels=Road_surface_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Gender distribution of driver')
plt.roatation=45
plt.axis('equal')
plt.show()
```

```python
sns.violinplot(data=data,x='Road_surface_conditions',y='Accident_severity')
plt.title('Severity Distribution of Accidents by Road surface condtions')
plt.xlabel('Road surface conditions')
plt.ylabel('Severity')
plt.grid(True)
plt.show()
```
